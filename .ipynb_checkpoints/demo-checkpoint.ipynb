```python
def takecommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("listening...")
        r.pause_threshold = 1
        audio = r.listen(source,timeout=2, phrase_time_limit=10)
    try:
        print("Recognizing...")
        query = r.recognize_google(audio,language='en-in')
        speak(f"I think you said {query}")
        # speak("if i am wrong, please say repeat")
        # audio2 = r.listen(source,timeout=2,phrase_time_limit=10)
        # query2 = r.recognize_google(audio,language='en-in')
        # if query2 == "repeat":
        #     takecommand()
            
    except Exception as e:
        speak("Please speak again")
        return None
    return query
```

```python
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voices',voices[0].id)
```

```python
# text to speech
def speak(text):
    engine.say(text)
    print(text)
    engine.runAndWait()
```

```python
if __name__ == '__main__':
    # speak("hello")
    # pass
    q=query.replace(" ","_")
    url = f"https://en.wikipedia.org/wiki/{q}"
    html = urlopen(url).read()
    soup = BeautifulSoup(html, features="html.parser")

    # kill all script and style elements
    for script in soup(["script", "style"]):
        script.extract()    # rip it out

    # get text
    text = soup.get_text()

    # break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
    # break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    # drop blank lines
    text = '\n'.join(chunk for chunk in chunks if chunk)

    
    speak(text)
```
